```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
f = open("C:/Users/ghosh/OneDrive/Desktop/Ram-Scion of Ikshvaku.txt", encoding="ISO-8859-1")
```

```python
raw_BOM= f.read()
FindWordsText_BOM = nltk.word_tokenize(raw_BOM) # Seperate text into words and preprocessing
text_BOM = nltk.Text(FindWordsText_BOM) # Spit out finished, searchable version of text
```

```python
print(len(text_BOM))
```

```python
print(text_BOM.count("Ram"))
```

```python
print(text_BOM.concordance("Ram"))
```

```python
text_BOM.common_contexts(["Sita", "Ram"])
```

```python
text_BOM.dispersion_plot(["Sita", "Ram", "Vishnu", "Vishwamitra", "Lakshman", "Ayodhya", "Mithila", "Urmila"])
```

```python
freq_dist=nltk.FreqDist(text_BOM)
freq_dist.most_common(25)
```
